### Replacing all null values with the mean of each column and removing all duplicates

We can fill null values with the mean, median or model of the respective column using the fillna() method. This is a common approach and is useful when there are only a few null values in the dataset just like this dataset which we will be using the mean value and the drop_duplicate
() method drop all duplicates in the dataset.


```python
dt = dt.fillna(dt_mean_values).drop_duplicates()
```

```python
meanFA=dt.fixed_acidity.mean()
meanVA=dt.volatile_acidity.mean()
meanCA=dt.citric_acid.mean()
meanRS=dt.residual_sugar.mean()
meanC=dt.chlorides.mean()
meanFSD=dt.free_sulfur_dioxide.mean()
meanTSD=dt.total_sulfur_dioxide.mean()
meanD=dt.density.mean()
meanPH=dt.pH.mean()
meanS=dt.sulphates.mean()
meanA=dt.alcohol.mean()
meanQ=dt.quality.mean()


```

### Pie chart of the columns in the dataset

```python
# Creating a pie chart  for our cleaned up data
colors = ['orange', 'violet', 'yellow', 'brown', 'red', 'purple',
          'pink', 'black', 'gray', 'blue', 'violet', 'cyan']
explode = (0.5,0.5,1,1,1,0,0,0.5,0,0.5,0,0.5)
meanValues = [meanFA, meanVA, meanCA, meanRS, meanC,
              meanFSD, meanTSD, meanD, meanPH, meanS, meanA, meanQ]
meanTopics = ['FA',
              'VA',
              'CA',
              'RS',
              'C',
              'FSD',
              'TSD',
              'density',
              'pH',
              'sulphates',
              'alcohol',
              'quality']



labels = meanTopics
sizes = meanValues

legendLabels = ['fixed_acidity (FA)',
              'volatile_acidity (VA)',
              'citric_acid (CA)',
              'residual_sugar (RS)',
              'chlorides (C)',
              'free_sulfur_dioxide (FSD)',
              'total_sulfur_dioxide (TSD)',
              'density',
              'pH',
              'sulphates',
              'alcohol',
              'quality']
```

```python
# plotting the pie chart
plt.pie(sizes, labels=labels, explode=explode,
        autopct='%1.1f%%', pctdistance=1.01, shadow=True, colors=colors, labeldistance=1.2)
plt.axis('equal')
plt.legend(title='index',labels=legendLabels)
plt.title('Mean Values for each of the columns in a pie chart')
plt.figure(figsize=(25,25))
plt.show()
```

### Bar graph of the columns in the dataset

```python
xtitle = 'Column names'
ytitle = 'Mean Values'
legendLabels = ['fixed_acidity (FA)',
              'volatile_acidity (VA)',
              'citric_acid (CA)',
              'residual_sugar (RS)',
              'chlorides (C)',
              'free_sulfur_dioxide (FSD)',
              'total_sulfur_dioxide (TSD)',
              'density',
              'pH',
              'sulphates',
              'alcohol',
              'quality']
```

```python
labels = meanTopics
sizes = meanValues
```

```python
# plotting a bar graph using the means as our data

fig = plt.figure(figsize=(10, 5))

plt.bar(labels, sizes, width=0.5, color=colors)
plt.xlabel(xtitle)
plt.ylabel(ytitle)
plt.title('Comparing mean of columns in a bar chart')
plt.legend( loc='upper right', )
plt.show()
```
